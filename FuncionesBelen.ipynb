```python
def columnas_a_minusculas(df):
    """
    Cambia los nombres de las columnas a minusculas
    parametros:
    dataframe del que sacamos las columnas
    diccionario_columnas:Diccionario que mapea nombres de columnas a sus nuevos nombres
    return:
    dataframe modificado
    """
    df.columns = map(str.lower, df.columns)
    df.columns

    return df
```

```python
def categorizar_department_id(df):
    """
    Esta función las siguientes columnas del dataframe() asigna valores
    transformados 
    Parámetros:
    Recibe un dataframe
    Returns:
    Devuelve el dataframe modificado
    """
    mapping = {
        1: 'Admin Offices',
        2: 'Executive Officer',
        3: 'IT/IS',
        4: 'Software Engineering',
        5: 'Production',
        6: 'Sales'
    }
    df['deptid'] = df['deptid'].map(mapping)
    return df
```

```python
def renombrar_columna(df, nombre_anterior, nombre_nuevo):
    """
    Renombra una columna en el DataFrame
    Parametros:
    - df: DataFrame
    - nombre_anterior: Nombre actual de la columna que va a ser renombrada.
    - nombre_nuevo: Nombre nuevo de la columna.

    Returns:
    - DataFrame actualizado.
    """
    if nombre_anterior in df.columns:   
        df.rename(columns={nombre_anterior: nombre_nuevo}, inplace=True)
        return df
    else:
        return df
    



```

```python
df = renombrar_columna(df,'dob','date_of_birth')
df = renombrar_columna(df,'department','xdepartment')
```

```python
def exploracion_inicial(df,lista_columnas):
    """Esta funcion hace una exploracion inicial de las columnas de un dataframe
        Parametros:
        df ... dataframe 
        lista_columnas .. lista de columnas del dataframe que hay que analizar
        return:
        nada, solamente lo analizamos y lo printeamos por pantalla para ir tomando decisiones"""
    for columna in lista_columnas:
        print('*********************************')
        print(f'Analizando la columna {columna}:')
        print(f'La cantidad de nulos es : {df[columna].isnull().sum()}')
        print(f'La columna es de tipo {df[columna].dtypes}')
        
        print(f'El porcentaje de nulos de esta columna es de  {df[columna].isnull().sum() / df.shape[0] * 100} %')
        print(f'Los valores unicos son: {df[columna].value_counts()}')
        print(f'Los composicion de la columna por valores es : { df[columna].value_counts()}')
```
