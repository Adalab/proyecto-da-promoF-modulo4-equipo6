```python
def info_columna(columna):
    print(f"Los valores que encontramos en la columna {columna.upper()} son {df_rh[columna].unique()}")
    print("-------------------------------------------------")
    print(f"De cada valos tenemos las siguientes cantidades:\n {df_rh[columna].value_counts()}")
    print("-------------------------------------------------")
    print(f"Y tenemos {df_rh[columna].isna().sum()} nulos")
```
